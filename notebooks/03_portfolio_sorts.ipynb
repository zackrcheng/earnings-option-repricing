# 03 — Nonparametric Portfolio Sorts

Univariate quintile sorts testing H1 (PreSkew → SR) and H2 (LN_VIX_pre → UR,
PreIV → UR_norm).  Means and Newey-West t-statistics are computed from the
time-series of cross-sectional portfolio averages.

```python
import warnings
warnings.filterwarnings('ignore')
import sys
from pathlib import Path
sys.path.insert(0, str(Path('..').resolve()))

from src.config import DATA_PATH, OUT_DIR, PRE_DAY, POST_DAY
from src.data_utils import load_panel, build_event_level_dataset
from src.portfolio import portfolio_sort_table

TABLES = OUT_DIR / 'tables'
TABLES.mkdir(parents=True, exist_ok=True)

df = load_panel(DATA_PATH)
event, _ = build_event_level_dataset(df, pre_day=PRE_DAY, post_day=POST_DAY)
print('Event-level observations:', len(event))
```

## H1 — PreSkew quintiles → Skew Revision (SR)

```python
h1_sort = portfolio_sort_table(event, 'PreSkew', 'SR', nport=5)
h1_sort.to_csv(TABLES / 'portfolio_sort_H1_PreSkew_to_SR.csv', index=False)
print('Saved: portfolio_sort_H1_PreSkew_to_SR.csv')
h1_sort
```

## H2a — LN_VIX_pre quintiles → Uncertainty Revision (UR)

```python
h2_vix = portfolio_sort_table(
    event, 'LN_VIX_pre', 'UR', nport=5, date_col='event_month'
)
h2_vix.to_csv(TABLES / 'portfolio_sort_H2_VIX_to_UR.csv', index=False)
print('Saved: portfolio_sort_H2_VIX_to_UR.csv')
h2_vix
```

## H2b — PreIV quintiles → UR_norm (uncertainty revision normalised by pre-event IV)

```python
h2_preiv = portfolio_sort_table(event, 'PreIV', 'UR_norm_by_PreIV', nport=5)
h2_preiv.to_csv(TABLES / 'portfolio_sort_H2_PreIV_to_URnorm.csv', index=False)
print('Saved: portfolio_sort_H2_PreIV_to_URnorm.csv')
h2_preiv
```
